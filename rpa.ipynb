### 엑셀에서 데이터 다루기

```python
import win32com.client
```

- 엑셀 프로그램 실행

```python
# 몇 월인지 확인
mon = input('몇월입니까?')
```

```python
excel = win32com.client.Dispatch('Excel.Application')
excel.Visible = True
```

- 워크북/시트 불러오기

```python
file_name = 'D:/LAB/RPA/data.xlsx'
wb = excel.Workbooks.Open(file_name)
ws = wb.Worksheets(mon + '월 판매(계획비)')
```

- 데이터 추출

```python
k6 = ws.Cells(13, 11).Value
print(k6)
```

```python
s6 = ws.Range("k14").Value
print(s6)
```

### 파워포인트에 텍스트 입력

```python
from pptx import Presentation

pptx_path = r"temp.pptx"

prs = Presentation(pptx_path)

slide = prs.slides[0]

# TextBox 18 찾기
for shape in slide.shapes:
    if shape.name == "TextBox 18":
        
        # 기존 텍스트 변경
        shape.text = "새로운 텍스트 입력"

        # 또는
        # shape.text_frame.text = "새로운 텍스트 입력"

        print("입력 완료")

# 저장
prs.save(pptx_path)
```

### 파워포인트에 데이터 입력

```python
from pptx import Presentation

pptx_path = r"temp.pptx"

prs = Presentation(pptx_path)

slide = prs.slides[0]

# "표 38" 객체 찾기
table_38_count = 0

for shape in slide.shapes:

    if shape.name == "표 38":

        if shape.has_table:
            table_38_count += 1

            table = shape.table

            # 2행 2열 값 입력
            # index는 0부터 시작
            if table_38_count == 1:
                table.cell(4, 2).text = str(k6)
                print("첫 번째 표 입력 완료")
            elif table_38_count == 2:
                table.cell(4, 2).text = str(s6)
                print("두 번째 표 입력 완료")
                break


# 저장
#prs.save(pptx_path)
prs.save(r"result.pptx")
```
